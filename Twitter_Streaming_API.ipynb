```python
import time
import datetime
import json
import pandas as pd
import numpy as np

from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
```

#### Define a class for saving the stream data and a basic error check

- This is a basic listener that just prints received tweets to stdout

```python
class StdoutListener(StreamListener):
    
    def on_data(self,data):
        try:
            print(data)
            savefile=open("./live_stream_tweets.json","a")
            savefile.write(data)
            savefile.write("\n")
            savefile.close()
            return true
        except BaseException:
            print("Failed on reading data")
            time.sleep(5)
            
    def on_error(self,status):
        print(status)
        
```

####  The below code handles the Twitter authetication and the connection to Twitter Streaming API

```python
if __name__== '__main__':
    l=StdoutListener()
    auth=OAuthHandler(consumer_key,consumer_secret)
    auth.set_access_token(access_token,access_token_secret)
    stream=Stream(auth,l)
    
    #This line filters the Twitter Streams to capture data by the keywords
    stream.filter(track=['bitcoin','btc','cryptocurrency','blockchain'])
```

#### Reading and storing in a Dataframe
- Lets read the the saved file as defined above in the location "./live_stream_tweets.json" in a read mode line by line.

```python
# Initialise an empyt list to store the tweets as they get read line by line
tweets_data = []

# open the stored file in a read mode\n",
tweets_file = open('./live_stream_tweets.json',"r")
#tweets_file

# for loop to iterate over the tweets_file and using json.loads\n",
# json.load is for a file and json.loads is for a string, here will use loads as we have already read the file above.\n",

for line in tweets_file:
    try:
        #print(line)
        tweet = json.loads(line)
        #print(tweet)
        tweets_data.append(tweet)
    except:
        continue
```

#### Build a DataFrame of tweet texts and other columns
----
- The columns I need in my dataframe are : creaed_at, id, text, lang
- *lang column will be dropped after filtering only english language tweets*
- Tweets length can be derived from the tweets text length which is max 140 characters.
- Next, is the spearation of date and time, as it would be useful in doing time-series like counts of tweets in a day or an hour.
- Lastly, reset the index after the english tweets filtering and drop the index and language column."

```python
def tweets_to_data_frame(tweets):
    df = pd.DataFrame(data=tweets, columns=['created_at','id','text','lang'])
    # Adding another column for the length of the tweet
    #df['Tweet_len'] = np.array([len(x) for x in df.text])
    df = df[df.lang == 'en']
    df['created_at'] = pd.to_datetime(df['created_at'])
    df['timestamp'] = pd.to_datetime(df['created_at'])
    df['created_at'] = df['created_at'].astype('object')
    df['Date'] = df['created_at'].dt.date
    df['Timestamp'] = df['timestamp'].dt.time
    df.reset_index()
    df.drop(['lang'], axis=1,inplace=True)
    df.drop(['index'], axis=1,inplace=True) 
    
```
